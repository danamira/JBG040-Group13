```python
# IMPORTS
from image_dataset import ImageDataset
from resnet import ResNet, Bottleneck
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
# Load Data
trainData = ImageDataset(Path("data/X_train.npy"), Path("data/Y_train.npy"))
testData = ImageDataset(Path("data/X_test.npy"), Path("data/Y_test.npy"))
```

```python
dataPointClasses = [i[1] for i in trainData]
```

```python
diseases = ['Atelectasis', 'Effusion', 'Infiltration', 'No finding', 'Nodule', 'Pneumothorax']
```

```python
trainDataPointClassLabels = [diseases[i] for i in dataPointClasses]
```

```python
# Explore the distributino of train data
plt.hist(trainDataPointClassLabels,label=diseases,bins=11)
plt.title("Distribution of clases in train data")
plt.show()
```

```python
trainCountDictionary = {x:0 for x in diseases}
for j in trainDataPointClassLabels:
    trainCountDictionary[j]+=1
```

```python
trainCountDictionary
```

```python
totalTrainDatapoints = len(trainData)
trainDataRelativeFrequency = {x:round(trainCountDictionary[x]/totalTrainDatapoints,2) for x in trainCountDictionary}
```

```python
trainDataRelativeFrequency
```

```python
trainDataRelativeFrequency
```

## Analysis of current model

```python
testDataPointClasses = [i[1] for i in testData]
```

```python
testDataPointClassLabels = [diseases[i] for i in testDataPointClasses]
```

```python
# Explore the distributino of train data
plt.hist(testDataPointClassLabels,label=diseases,bins=11)
plt.title("Distribution of clases in test data")
plt.show()
```

```python
 modelPath = f"model_weights/best.txt"
```

```python
model = ResNet(Bottleneck, layer_list=[1, 3, 4, 2, 1], num_classes=6, num_channels=1)
model.load_state_dict(torch.load(modelPath, map_location=torch.device('mps')))
```

```python
xTest=torch.from_numpy(testData.imgs)
yTest=torch.from_numpy(testData.targets)
```

```python
xTest=xTest[0:500]
yTest=yTest[0:500]
```

```python
xTest=xTest.to('mps').float()
```

```python
model.to('mps')
results=model(xTest)
```

```python
predictionsOnTestData=torch.argmax(results,dim=1)
predictedLabelOnTestData = [diseases[x] for x in predictionsOnTestData.cpu().numpy()]
```

```python
yTestLabels = [diseases[x] for x in yTest]
```

```python
plt.hist(yTestLabels,bins=11,label='actual')
plt.hist(predictedLabelOnTestData,bins=11,alpha=0.5,label='predictions')
plt.title('Frequency of classes and predictions based on test data')
plt.legend()
```

<h2>Testing batch sampler</h2>
<p>Since there seems to be a meaningful underreprestation of the disease 1 (infiltration) maybe the data provided to the model during the training is missing something. We shall check this in the following blocks of code.</p>

```python
# IMPORT BATCH SAMPLER 
from batch_sampler import BatchSampler
from tqdm import tqdm
```

```python
trainSampler = BatchSampler(batch_size=500, dataset=trainData, balanced=True) 
## Batches of size 25 and equal number of diseases (balanced parameter)
```

```python
for x in trainSampler:
    a= x
    break
```

```python
firstTrainBatchClasses=a[1]
firstTrainBatchClassLabels = [diseases[i] for i in firstTrainBatchClasses]
```

```python
plt.hist(firstTrainBatchClassLabels,bins=11)
plt.title('Distribution of classes in a single training batch')
plt.show()
```

<p><b>Result:</b> The balanced parameter seems to works just fine. The training data points of each class provided by the sample seem to have roughly equal amount in a single batch.</p>
